```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import ta
import numpy as np
pd.options.mode.chained_assignment = None
```

```python
tickers = pd.read_html("https://en.wikipedia.org/wiki/NIFTY_50", match="Company Name")[0]
```

```python
tickers['Symbol'] = tickers['Symbol'].astype(str) + '.NS'
tickers
```

```python
symbols = tickers.Symbol.to_list()
```

```python
symbols
```

```python
df = yf.download(symbols[36], start='2010-01-01')
```

```python
df
```

```python
def tacalc(df):
    df['sma200'] = ta.trend.sma_indicator(df.Close, window=200)
    df['rsi'] = ta.momentum.rsi(df.Close, window=10)
    df['signal'] = np.where((df.Close > df.sma200) & (df.rsi<30), True, False)
```

```python
def getactuals(df):
    buying_dates = []
    selling_dates = []

    for i in range(len(df)-11):
        if df.signal.iloc[i]:
            buying_dates.append(df.iloc[i+1].name)
            for j in range(1,11):
                if df['rsi'].iloc[i+j] > 40:
                    selling_dates.append(df.iloc[i+j+1].name)
                    break
                elif j==10:
                    selling_dates.append(df.iloc[i+j+1].name)
                    
    frame = pd.DataFrame({'buying_dates':buying_dates, 'selling_dates':selling_dates})
    actuals = frame[frame.buying_dates > frame.selling_dates.shift(1)]
    actuals = frame[:1].append(actuals)
                    
    return actuals
```

```python
tacalc(df)
actuals = getactuals(df)
```

```python
profits = \
(df.loc[actuals.selling_dates].Open.values - df.loc[actuals.buying_dates].Open.values)/df.loc[actuals.buying_dates].Open.values
```

```python
profits
```

```python
sum(profits)
```

# Test on all companies

```python
matrixprofits = []

for i in range(len(symbols)):
    df = yf.download(symbols[i], start='2010-01-01')
    tacalc(df)
    actualtrades = getactuals(df)
    profits = \
    (df.loc[actualtrades.selling_dates].Open.values - df.loc[actualtrades.buying_dates].Open.values)/df.loc[actualtrades.buying_dates].Open.values
    matrixprofits.append(profits)
```

```python
s = 0
for element in matrixprofits:
    s += sum(element)
```
